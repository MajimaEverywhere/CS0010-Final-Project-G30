```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
def filter_less_than(grouped, threshold):
    if len(grouped) < threshold:
        return False
    else:
        return True
signs = pd.read_csv("cityTrafficSigns.tsv", sep="\t")
#Initialiing the sign dataset along with implementing a size filter.
```

```python
signs["neighborhood"].value_counts().sort_values(ascending=True)
```

```python

stopSign = signs
stopSign = stopSign[stopSign['description'].str.contains("Stop", na = False)]
stopSign = stopSign.groupby("neighborhood").filter(filter_less_than, threshold = 300)
stopSign["neighborhood"].value_counts().sort_index().plot(kind = 'bar')
#Dataset is seeing how many stop signs are witnin every neighborhood. Only the signs
```

```python
sSFlats = signs
sSFlats = sSFlats[sSFlats['neighborhood'].str.contains("South Side Flats", na = False)]
sSFlats = sSFlats.groupby("description").filter(filter_less_than, threshold = 100)
sSFlats["description"].value_counts().sort_index().plot(kind = 'bar')
```
